```python

# read chosen-fields file 
ndf = pd.read_csv('chosen-fields.csv')

# Add new field of Prospecct_Sum for women between 15 to 44
ndf['Prospecct_Sum'] = ndf['여자15세부터19세생활인구수'] + ndf['여자20세부터24세생활인구수'] + ndf['여자30세부터34세생활인구수'] + ndf['여자35세부터39세생활인구수'] + ndf['여자40세부터44세생활인구수']

ndf
```

```python
# select only 기준일', '시간대구분', '행정동코드', '총생활인구수', '행정동코드1', 'sum'
adf = ndf[['기준일', '시간대구분', '총생활인구수', '행정동코드', 'Prospecct_Sum']]
adf
```

```python
# filters data during 0 am - 6am and after 10 pm
tdf = adf[(adf['시간대구분']>=7) & (adf['시간대구분'] < 23)]
tdf
```

```python
# group the same towns and calculate the mean number of Prospect_Sum of the towns 
# gdf = tdf.groupby('행정동코드')
tdf['Prospecct_Sum-Mean'] = tdf.groupby('행정동코드')['Prospecct_Sum'].transform('mean')
```

```python
# group by the towns again to get the described data 
gdf = tdf.groupby('행정동코드')
kdf = gdf.describe()
```

```python
# Top 5 towns among the data 

sdf = kdf['Prospecct_Sum-Mean']
fdf = sdf.sort_values(by=['mean'], ascending=[False]).head(5)
fdf
```
